```python
import os
from PIL import Image, ImageDraw
import rasterio
import geopandas as gpd
import shutil
from tqdm import tqdm
from shapely.geometry import mapping
from affine import Affine

image_dir = '/content/drive/MyDrive/CIL_DATA/SN5_roads_train_AOI_7_Moscow/nfs/data/cosmiq/spacenet/competitions/SN5_roads/tiles_upload/train/AOI_7_Moscow/PS-RGB'
geojson_dir = '/content/drive/MyDrive/CIL_DATA/SN5_roads_train_AOI_7_Moscow/nfs/data/cosmiq/spacenet/competitions/SN5_roads/tiles_upload/train/AOI_7_Moscow/geojson_roads_speed'
output_dir = '/content/drive/MyDrive/CIL_DATA/processed_data2'

os.makedirs(output_dir, exist_ok=True)

image_files = [f for f in os.listdir(image_dir) if f.endswith('.tif')]
processed_count = 0

def geojson_to_mask(geojson_path, image_size, transform):
    gdf = gpd.read_file(geojson_path)
    mask = Image.new('L', image_size, 0)
    draw = ImageDraw.Draw(mask)

    for _, row in gdf.iterrows():
        geom = row.geometry
        if geom is not None:
            if geom.type == 'LineString':
                coordinates = [transform * coord for coord in geom.coords]
                coordinates = [(int(x), int(y)) for x, y in coordinates if 0 <= x < image_size[0] and 0 <= y < image_size[1]]  # Ensure coordinates are within image bounds
                if coordinates:
                    draw.line(coordinates, fill=255, width=12)  # Adjust width as needed for visibility
            elif geom.type == 'MultiLineString':
                for line in geom.geoms:
                    coordinates = [transform * coord for coord in line.coords]
                    coordinates = [(int(x), int(y)) for x, y in coordinates if 0 <= x < image_size[0] and 0 <= y < image_size[1]]  # Ensure coordinates are within image bounds
                    if coordinates:
                        draw.line(coordinates, fill=255, width=5)  # Adjust width as needed for visibility

    return mask

with tqdm(total=5000, desc="Processing images", unit="file") as pbar:
    for image_file in image_files:
        if processed_count >= 5000:
            break

        image_path = os.path.join(image_dir, image_file)

        with rasterio.open(image_path) as src:
            img_array = src.read()
            img = Image.fromarray(img_array.transpose((1, 2, 0)))

            transform = ~src.transform

            geojson_file = image_file.replace('PS-RGB', 'geojson_roads_speed').replace('.tif', '.geojson')
            geojson_path = os.path.join(geojson_dir, geojson_file)

            if os.path.exists(geojson_path):
                mask = geojson_to_mask(geojson_path, (src.width, src.height), transform)

                img_resized = img.resize((400, 400), Image.ANTIALIAS)
                mask_resized = mask.resize((400, 400), Image.NEAREST)

                img_resized.save(os.path.join(output_dir, image_file.replace('.tif', '.png')))
                mask_resized.save(os.path.join(output_dir, image_file.replace('.tif', '_mask.png')))

        processed_count += 1
        pbar.update(1)

print(f'Processed {processed_count} images and their corresponding geojson files.')
```

## OTTAWA Dataset


```python
import os
from PIL import Image
import rasterio
import numpy as np

dataset_dir = '/content/drive/MyDrive/CIL_DATA/Ottawa-Dataset'
output_dir = '/content/drive/MyDrive/CIL_DATA/processed_Ottawa_data'

os.makedirs(output_dir, exist_ok=True)

for folder_name in os.listdir(dataset_dir):
    folder_path = os.path.join(dataset_dir, folder_name)
    if os.path.isdir(folder_path):
        for file_name in os.listdir(folder_path):
            if file_name.endswith('.tif'):
                tif_path = os.path.join(folder_path, file_name)
                mask_path = os.path.join(folder_path, 'segmentation.png')

                if not os.path.exists(mask_path):
                    print(f"Mask not found for {tif_path}. Skipping.")
                    continue

                with rasterio.open(tif_path) as src:
                    img_array = src.read((1, 2, 3))
                    img_array = np.transpose(img_array, (1, 2, 0))
                    img_height, img_width, _ = img_array.shape

                mask = Image.open(mask_path)
                mask = mask.convert('L')
                mask_array = np.array(mask)
                mask_height, mask_width = mask_array.shape

                if img_height != mask_height or img_width != mask_width:
                    print(f"Image and mask dimensions do not match for {tif_path}. Skipping.")
                    continue

                threshold = np.mean(mask_array)
                inverted_mask = (mask_array <= threshold).astype(np.uint8) * 255

                segment_count = 0
                for y in range(0, img_height, 400):
                    for x in range(0, img_width, 400):
                        if x + 400 <= img_width and y + 400 <= img_height:
                            img_segment = img_array[y:y+400, x:x+400]
                            mask_segment = inverted_mask[y:y+400, x:x+400]

                            img_segment_pil = Image.fromarray(img_segment)
                            mask_segment_pil = Image.fromarray(mask_segment)

                            img_segment_pil.save(os.path.join(output_dir, f'{file_name.replace(".tif", "")}_segment_{segment_count}.png'))
                            mask_segment_pil.save(os.path.join(output_dir, f'{file_name.replace(".tif", "")}_segment_{segment_count}_mask.png'))

                            segment_count += 1

print('Segments have been extracted and saved.')
```

```python
import os
from PIL import Image
from tqdm import tqdm
import shutil
from sklearn.model_selection import train_test_split
import numpy as np

data_dir = "/content/drive/MyDrive/CIL_DATA/archive/train"
output_dir = "/content/drive/MyDrive/CIL_DATA/GEO_processed_data_final"
train_image_dir = os.path.join(output_dir, "train/images")
train_label_dir = os.path.join(output_dir, "train/labels")
val_image_dir = os.path.join(output_dir, "val/images")
val_label_dir = os.path.join(output_dir, "val/labels")

os.makedirs(train_image_dir, exist_ok=True)
os.makedirs(train_label_dir, exist_ok=True)
os.makedirs(val_image_dir, exist_ok=True)
os.makedirs(val_label_dir, exist_ok=True)

N = 99999

image_files = [f for f in os.listdir(data_dir) if f.endswith('_sat.jpg')]
label_files = [f for f in os.listdir(data_dir) if f.endswith('_mask.png')]

image_files.sort()
label_files.sort()

image_files = image_files[:N]
label_files = label_files[:N]

def has_non_black_pixels(image):
    return np.any(np.array(image) > 0)

def crop_image(image, mask, crop_size):
    width, height = image.size
    new_width, new_height = crop_size
    attempts = [
        ((width - new_width) // 2, (height - new_height) // 2),
        (0, 0),
        (width - new_width, 0),
        (0, height - new_height),
        (width - new_width, height - new_height)
    ]

    for left, top in attempts:
        right = left + new_width
        bottom = top + new_height
        cropped_mask = mask.crop((left, top, right, bottom))
        if has_non_black_pixels(cropped_mask):
            return image.crop((left, top, right, bottom)), cropped_mask

    left, top = attempts[0]
    right = left + new_width
    bottom = top + new_height
    return image.crop((left, top, right, bottom)), mask.crop((left, top, right, bottom))

temp_image_dir = os.path.join(output_dir, "temp/images")
temp_label_dir = os.path.join(output_dir, "temp/labels")

os.makedirs(temp_image_dir, exist_ok=True)
os.makedirs(temp_label_dir, exist_ok=True)

for img_file, lbl_file in tqdm(zip(image_files, label_files), total=len(image_files)):
    img_path = os.path.join(data_dir, img_file)
    lbl_path = os.path.join(data_dir, lbl_file)

    img = Image.open(img_path)
    lbl = Image.open(lbl_path)

    img_cropped, lbl_cropped = crop_image(img, lbl, (400, 400))

    img_cropped.save(os.path.join(temp_image_dir, img_file))
    lbl_cropped.save(os.path.join(temp_label_dir, lbl_file))

temp_image_files = os.listdir(temp_image_dir)
train_files, val_files = train_test_split(temp_image_files, test_size=0.2, random_state=42)

for file in train_files:
    img_src = os.path.join(temp_image_dir, file)
    lbl_src = os.path.join(temp_label_dir, file.replace('_sat.jpg', '_mask.png'))
    img_dst = os.path.join(train_image_dir, file)
    lbl_dst = os.path.join(train_label_dir, file.replace('_sat.jpg', '_mask.png'))

    if os.path.exists(img_src) and os.path.exists(lbl_src):
        shutil.move(img_src, img_dst)
        shutil.move(lbl_src, lbl_dst)

for file in val_files:
    img_src = os.path.join(temp_image_dir, file)
    lbl_src = os.path.join(temp_label_dir, file.replace('_sat.jpg', '_mask.png'))
    img_dst = os.path.join(val_image_dir, file)
    lbl_dst = os.path.join(val_label_dir, file.replace('_sat.jpg', '_mask.png'))

    if os.path.exists(img_src) and os.path.exists(lbl_src):
        shutil.move(img_src, img_dst)
        shutil.move(lbl_src, lbl_dst)

shutil.rmtree(temp_image_dir)
shutil.rmtree(temp_label_dir)
```

```python
import os
import shutil
from tqdm import tqdm
from PIL import Image

source_dirs = {
    "GEO": {
        "path": "/content/drive/MyDrive/CIL_DATA/GEO_processed_data_final",
        "image_suffix": "_sat.jpg",
        "mask_suffix": "_mask.png"
    },
    "Ottawa": {
        "path": "/content/drive/MyDrive/CIL_DATA/ottawa_final",
        "image_suffix": ".png",
        "mask_suffix": "_mask.png"
    }
}

destination_dir = "/content/drive/MyDrive/CIL_DATA/CombinedDatasets_Smaller"

train_image_dir = os.path.join(destination_dir, "train/images")
train_label_dir = os.path.join(destination_dir, "train/labels")
val_image_dir = os.path.join(destination_dir, "val/images")
val_label_dir = os.path.join(destination_dir, "val/labels")

os.makedirs(train_image_dir, exist_ok=True)
os.makedirs(train_label_dir, exist_ok=True)
os.makedirs(val_image_dir, exist_ok=True)
os.makedirs(val_label_dir, exist_ok=True)

def copy_files_with_progress(src_image_dir, src_label_dir, dst_image_dir, dst_label_dir, image_suffix, mask_suffix, start_id=0):
    image_files = sorted([f for f in os.listdir(src_image_dir) if f.endswith(image_suffix)])
    label_files = sorted([f for f in os.listdir(src_label_dir) if f.endswith(mask_suffix)])

    id_counter = start_id

    for image_file in tqdm(image_files, desc=f"Copying from {src_image_dir}", unit="file"):
        base_name = image_file.replace(image_suffix, "")
        label_file = f"{base_name}{mask_suffix}"

        if label_file not in label_files:
            print(f"Warning: No matching label for {image_file}")
            continue

        new_image_file = f"{id_counter:06d}.png"
        new_label_file = f"{id_counter:06d}.png"

        src_image_path = os.path.join(src_image_dir, image_file)
        src_label_path = os.path.join(src_label_dir, label_file)
        dst_image_path = os.path.join(dst_image_dir, new_image_file)
        dst_label_path = os.path.join(dst_label_dir, new_label_file)

        with Image.open(src_image_path) as img:
            img.save(dst_image_path, 'PNG')

        shutil.copy(src_label_path, dst_label_path)

        id_counter += 1

    return id_counter

next_id = 0
for key, value in source_dirs.items():
    next_id = copy_files_with_progress(
        os.path.join(value["path"], "train/images"),
        os.path.join(value["path"], "train/labels"),
        train_image_dir,
        train_label_dir,
        value["image_suffix"],
        value["mask_suffix"],
        next_id
    )

next_id = 0
for key, value in source_dirs.items():
    next_id = copy_files_with_progress(
        os.path.join(value["path"], "val/images"),
        os.path.join(value["path"], "val/labels"),
        val_image_dir,
        val_label_dir,
        value["image_suffix"],
        value["mask_suffix"],
        next_id
    )

print("Data successfully combined into", destination_dir)

```

```python
import random
from PIL import Image
import matplotlib.pyplot as plt

def show_random_samples(image_dir, label_dir, num_samples=10):
    image_files = sorted([f for f in os.listdir(image_dir) if f.endswith('.png')])
    label_files = sorted([f for f in os.listdir(label_dir) if f.endswith('.png')])

    num_files = min(len(image_files), len(label_files))

    sampled_indices = random.sample(range(num_files), num_samples)

    plt.figure(figsize=(20, num_samples * 2))

    for i, idx in enumerate(sampled_indices):
        image_path = os.path.join(image_dir, image_files[idx])
        label_path = os.path.join(label_dir, label_files[idx])

        image = Image.open(image_path)
        label = Image.open(label_path)

        plt.subplot(num_samples, 2, 2 * i + 1)
        plt.imshow(image)
        plt.title(f'Image: {image_files[idx]}')
        plt.axis('off')

        plt.subplot(num_samples, 2, 2 * i + 2)
        plt.imshow(label, cmap='gray')
        plt.title(f'Mask: {label_files[idx]}')
        plt.axis('off')

    plt.tight_layout()
    plt.show()

show_random_samples(train_image_dir, train_label_dir)
```
